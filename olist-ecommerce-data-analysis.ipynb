# **Step 3 — Missing Values Check**

```python

# STEP 4: Fix Missing Values

# orders table mein jo orders approve nahi hue
# unka approved time purchase time se fill karte hain
orders['order_approved_at'].fillna(
    orders['order_purchase_timestamp'], inplace=True
)

# products mein jo category name nahi hai
# unhe unknown kar dete hain
products['product_category_name'].fillna('unknown', inplace=True)

# products ke numeric columns mein
# median se fill karte hain kyunki data skewed ho sakta hai
num_cols = [
    'product_name_lenght',
    'product_description_lenght', 
    'product_photos_qty',
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm'
]
for col in num_cols:
    products[col].fillna(products[col].median(), inplace=True)

# reviews mein comment nahi hain to
# simply No Title aur No Comment likh dete hain
reviews['review_comment_title'].fillna('No Title', inplace=True)
reviews['review_comment_message'].fillna('No Comment', inplace=True)

print("Missing values fix ho gaye!")
```
