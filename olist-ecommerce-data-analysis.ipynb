# **Step 3 — Missing Values Check**

```python

# Step 3 — Check Missing Values in All Datasets

datasets = {
    'orders'    : orders,
    'items'     : items,
    'customers' : customers,
    'products'  : products,
    'reviews'   : reviews,
    'sellers'   : sellers,
    'payments'  : payments
}

for name, df in datasets.items():
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    
    print(f"\n{name}:")
    if len(missing) == 0:
        print("  no missing values")
    else:
        for col, count in missing.items():
            pct = round((count / len(df)) * 100, 2)
            print(f"  {col}: {count} missing ({pct}%)")

```

# **Step 4 — Fix Missing Values**

```python
orders['order_approved_at'].fillna(
    orders['order_purchase_timestamp'], inplace=True
)
products['product_category_name'].fillna('unknown', inplace=True)
num_cols = [
    'product_name_lenght',
    'product_description_lenght',
    'product_photos_qty',
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm'
]
for col in num_cols:
    products[col].fillna(products[col].median(), inplace=True)
reviews['review_comment_title'].fillna('No Title', inplace=True)
reviews['review_comment_message'].fillna('No Comment', inplace=True)

print("done — missing values fixed!")
```

# **Step 5 — Duplicate Values Check**

```python
# Check Duplicate Values

for name, df in datasets.items():
    dups = df.duplicated().sum()
    print(f"{name}: {dups} duplicates")
```

# **Step 6 — Fix Data Types**

```python
date_cols_orders = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
for col in date_cols_orders:
    orders[col] = pd.to_datetime(orders[col], errors='coerce')

items['shipping_limit_date'] = pd.to_datetime(
    items['shipping_limit_date'], errors='coerce'
)
reviews['review_creation_date'] = pd.to_datetime(
    reviews['review_creation_date'], errors='coerce'
)
reviews['review_answer_timestamp'] = pd.to_datetime(
    reviews['review_answer_timestamp'], errors='coerce'
)
print("done — date columns converted!")
# verify
print("\norders date types:")
print(orders[date_cols_orders].dtypes)
```

# **Step 7 — Fix Text Issues**

```python
customers['customer_city'] = customers['customer_city'].str.title()
sellers['seller_city'] = sellers['seller_city'].str.title()
category_translation = pd.read_csv(
    path + 'product_category_name_translation.csv'
)
products = products.merge(
    category_translation,
    on='product_category_name',
    how='left'
)
products.drop(columns=['product_category_name'], inplace=True)
products.rename(columns={
    'product_category_name_english': 'product_category_name'
}, inplace=True)
products.rename(columns={
    'product_name_lenght'        : 'product_name_length',
    'product_description_lenght' : 'product_description_length'
}, inplace=True)

print("done — text issues fixed!")
# verify
print("\nproducts categories (sample):")
print(products['product_category_name'].unique()[:10])
```

# **Step 8 — Feature Engineering**

```python
orders['order_year']    = orders['order_purchase_timestamp'].dt.year
orders['order_month']   = orders['order_purchase_timestamp'].dt.month
orders['order_day']     = orders['order_purchase_timestamp'].dt.day_name()
orders['order_hour']    = orders['order_purchase_timestamp'].dt.hour

orders['delivery_days'] = (
    orders['order_delivered_customer_date'] -
    orders['order_purchase_timestamp']
).dt.days

orders['is_late'] = (
    orders['order_delivered_customer_date'] >
    orders['order_estimated_delivery_date']
)

print("done — new features created!")
# verify
print("\nsample:")
print(orders[['order_purchase_timestamp',
              'order_year', 'order_month',
              'order_day', 'delivery_days',
              'is_late']].head())
```

# **Step 9 — Final Check**

```python
for name, df in datasets.items():
    missing = df.isnull().sum().sum()
    dups    = df.duplicated().sum()
    print(f"{name}: {df.shape[0]} rows, {df.shape[1]} cols | missing: {missing} | duplicates: {dups}")

print("\norders new columns sample:")
print(orders[['order_year', 'order_month', 'order_day',
              'delivery_days', 'is_late']].describe())
```

# **Step 10 — Outliers Check**

```python
#  Check Outliers in delivery_days and price

# delivery days outliers
Q1 = orders['delivery_days'].quantile(0.25)
Q3 = orders['delivery_days'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = orders[
    (orders['delivery_days'] < lower) |
    (orders['delivery_days'] > upper)
]

print(f"delivery_days outliers: {len(outliers)}")
print(f"normal range: {lower:.0f} to {upper:.0f} days")

# price outliers
Q1 = items['price'].quantile(0.25)
Q3 = items['price'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_price = items[
    (items['price'] < lower) |
    (items['price'] > upper)
]

print(f"\nprice outliers: {len(outliers_price)}")
print(f"normal range: {lower:.0f} to {upper:.0f}")
print(f"max price in data: {items['price'].max()}")
```

# **Visualization — Chart 1**

```python
# Visualization — Chart 1
# Order Status Distribution

order_status = orders['order_status'].value_counts()

plt.figure(figsize=(10, 5))

sns.barplot(x=order_status.index,
            y=order_status.values,
            palette='Blues_d')

plt.title('Order Status Distribution', fontsize=15, fontweight='bold')
plt.xlabel('Status')
plt.ylabel('Number of Orders')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

print(order_status)
```

# **Chart 2 — Monthly Revenue Trend**

```python
# Visualization — Chart 2
# Monthly Revenue Trend

monthly_revenue = (
    orders.merge(items, on='order_id')
    .groupby(['order_year', 'order_month'])['price']
    .sum()
    .reset_index()
)


monthly_revenue = monthly_revenue[monthly_revenue['order_year'] != 2016]


monthly_revenue['date'] = pd.to_datetime(
    monthly_revenue[['order_year', 'order_month']].assign(day=1)
    .rename(columns={'order_year': 'year', 'order_month': 'month'})
)

plt.figure(figsize=(12, 5))

plt.plot(monthly_revenue['date'],
         monthly_revenue['price'],
         marker='o',
         linewidth=2,
         color='steelblue')

plt.fill_between(monthly_revenue['date'],
                 monthly_revenue['price'],
                 alpha=0.15,
                 color='steelblue')

plt.title('Monthly Revenue Trend', fontsize=15, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Revenue (BRL)')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

# **Chart 3 — Top 10 Product Categories**

```python

# Top 10 Product Categories by Orders

top_categories = (
    items.merge(products, on='product_id')
    .groupby('product_category_name')['order_id']
    .count()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)
top_categories.columns = ['category', 'orders']

plt.figure(figsize=(10, 6))

sns.barplot(data=top_categories,
            x='orders',
            y='category',
            palette='viridis')

plt.title('Top 10 Product Categories by Orders',
          fontsize=15, fontweight='bold')
plt.xlabel('Number of Orders')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```
